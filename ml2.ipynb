```python
# import dependencies
import pandas as pd
```

```python
# import the all_channels.csv file
random_channels_df = pd.read_csv('other_csvs/random_channels_df.csv')
random_comment_df = pd.read_csv('other_csvs/random_comment_df.csv')
# rename the column to match the comment_df
channels_df = random_channels_df.rename(columns={'id': 'channel_id'})

# merge the two dataframes
random_merged_df = pd.merge(channels_df, random_comment_df, on='channel_id')
```

```python
# drop the columns
random_merged_df = random_merged_df.drop(columns_to_drop, axis=1)
```
